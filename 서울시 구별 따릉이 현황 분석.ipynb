#### 서울시 구별 인구 데이터 다운로드

https://data.seoul.go.kr/dataList/419/S/2/datasetView.do



#### txt 파일로 내려받아 엑셀을 이용해 csv 파일로 변환



```python
# 서울시 구별 인구 데이터 읽기
pop_Seoul = pd.read_csv('./report.csv', 
                        header=2,
                        usecols= [1, 3],
                        encoding='cp949')
pop_Seoul.rename(columns={pop_Seoul.columns[1] : '인구수'}, inplace=True)

# 숫자 string -> int
pop_Seoul['인구수'] = pop_Seoul['인구수'].map(lambda x: x.replace(',', ''))
pop_Seoul['인구수'] = pop_Seoul['인구수'].astype(int)
pop_Seoul.head()
```

```python
# 따릉이 데이터와 인구 데이터 합치기
Bike_Seoul_Sum = Bike_Seoul_Sum.reset_index()
# Bike_Seoul_Sum.head()
Bike_Seoul_Sum.rename(columns={Bike_Seoul_Sum.columns[0] : '자치구'}, inplace=True)
# Bike_Seoul_Sum.head()
data_result = pd.merge(Bike_Seoul_Sum, pop_Seoul, on='자치구')
data_result.head()
```

```python
data_result['인구수 대비 거치대 비율'] = data_result['거치대수'] / data_result['인구수'] * 100
data_result.head()
# type(data_result['거치대수'][0])
# type(data_result['인구수'][0])
```

```python
data_result.set_index('자치구', inplace=True)
data_result = data_result.drop('합계')
data_result
```

```python
data_result['인구수 대비 거치대 비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['거치대수'], s=50)
plt.xlabel('인구수')
plt.ylabel('거치대수')
plt.grid()
plt.show()
```

```python
# 데이터를 대표하는 1차 함수 구하기
import numpy as np
fp1 = np.polyfit(data_result['인구수'], data_result['거치대수'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
# 경향성 파악하기
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['거치대수'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('거치대수')
plt.grid()
plt.show()
```

```python
# 오차에 따른 colormap 표시하기
fp1 = np.polyfit(data_result['인구수'], data_result['거치대수'], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

data_result['오차'] = np.abs(data_result['거치대수'] - f1(data_result['인구수']))

df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['거치대수'],
           c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['거치대수'][n]*0.98,
            df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('거치대수')

plt.colorbar()
plt.grid()
plt.show()
```
