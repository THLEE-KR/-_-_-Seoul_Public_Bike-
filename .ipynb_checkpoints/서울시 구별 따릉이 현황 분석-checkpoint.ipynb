```python
#     
pop_Seoul = pd.read_csv('./report.csv', 
                        header=2,
                        usecols= [1, 3],
                        encoding='cp949')
pop_Seoul.rename(columns={pop_Seoul.columns[1] : ''}, inplace=True)

#  string -> int
pop_Seoul[''] = pop_Seoul[''].map(lambda x: x.replace(',', ''))
pop_Seoul[''] = pop_Seoul[''].astype(int)
pop_Seoul.head()
```

```python
#     
Bike_Seoul_Sum = Bike_Seoul_Sum.reset_index()
# Bike_Seoul_Sum.head()
Bike_Seoul_Sum.rename(columns={Bike_Seoul_Sum.columns[0] : ''}, inplace=True)
# Bike_Seoul_Sum.head()
data_result = pd.merge(Bike_Seoul_Sum, pop_Seoul, on='')
data_result.head()
```

```python
data_result['   '] = data_result[''] / data_result[''] * 100
data_result.head()
# type(data_result[''][0])
# type(data_result[''][0])
```

```python
data_result.set_index('', inplace=True)
data_result = data_result.drop('')
data_result
```

```python
data_result['   '].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result[''], data_result[''], s=50)
plt.xlabel('')
plt.ylabel('')
plt.grid()
plt.show()
```

```python
#   1  
import numpy as np
fp1 = np.polyfit(data_result[''], data_result[''], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
#  
plt.figure(figsize=(10,10))
plt.scatter(data_result[''], data_result[''], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('')
plt.ylabel('')
plt.grid()
plt.show()
```

```python
#   colormap 
fp1 = np.polyfit(data_result[''], data_result[''], 1)

f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

data_result[''] = np.abs(data_result[''] - f1(data_result['']))

df_sort = data_result.sort_values(by='', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result[''], data_result[''],
           c=data_result[''], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

for n in range(10):
    plt.text(df_sort[''][n]*1.02, df_sort[''][n]*0.98,
            df_sort.index[n], fontsize=15)
    
plt.xlabel('')
plt.ylabel('')

plt.colorbar()
plt.grid()
plt.show()
```

```python
import json
import folium

#   json file 
geo_path = './02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

map = folium.Map(location=[37.5502, 126.982], zoom_start=10,
                tiles='Stamen Toner')
map.choropleth(geo_data = geo_str,
              data = df_sort['   '],
              columns = [df_sort.index, df_sort['   ']],
              fill_color = 'YlGnBu',
              key_on = 'feature.id')
map
```
